Purpose:
Clean and standardise workforce, performance, attendance, and sales data to enable reliable analysis and reporting.

```python
import pandas as pd
employees = pd.read_csv("../data/employees.csv")
departments = pd.read_csv("../data/departments.csv")
performance = pd.read_csv("../data/performance.csv")
attendance = pd.read_csv("../data/attendance.csv")
sales = pd.read_csv("../data/sales.csv")

employees["hire_date"] = pd.to_datetime(employees["hire_date"], dayfirst=True)
performance["review_date"] = pd.to_datetime(performance["review_date"], dayfirst=True)
sales["sale_date"] = pd.to_datetime(sales["sale_date"], dayfirst=True)

performance["performance_score"].value_counts(dropna=False) #identify invalid values
performance.loc[
    (performance["performance_score"] < 1) | 
    (performance["performance_score"] > 5),
    "performance_score"
] = None

attendance["month"] = attendance["month"].str.strip().str.lower()

attendance["month"] = attendance["month"].replace({
    "jan": "january",
    "feb": "february"
})
```
